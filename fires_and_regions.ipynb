## Following the same structure for each xlsx file
- creating a dataframe
- setting the first row as a header
- reseting the index
- turning the column "Ημερ/νία Έναρξης" as datetime type
- extracting the year
- summing the burned stremmata (which mean burned acres) for each category by year and district and calling it as a new dataframe
- and printing the final dataframe for the year

```python
#I am calling my excel and making it a dataframe
df = pd.read_excel("Dasikes_Pyrkagies_2000-2012.xlsx")
# I am setting the first row as header, because it wasn't like that when I tried it
df.columns = df.iloc[0]
df = df[1:]

# I am reseting the index
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
#I am making the column "Ημερ/νία Έναρξης" as datetime type
df["Ημερ/νία Έναρξης"] = pd.to_datetime(df["Ημερ/νία Έναρξης"])

# I am extracting the year from the column since that's what I will be working with
df["Έτος"] = df["Ημερ/νία Έναρξης"].dt.year

# I am summing the burned stremmata (which mean burned acres) for each category by year and district and calling it as a new dataframe
burned_stremmata_by_category = df.groupby(["Έτος", "Νομός"])[["Δάση", "Δασική Έκταση", "Άλση", "Χορτ/κές Εκτάσεις", "Καλάμια - Βάλτοι", "Γεωργικές Εκτάσεις", "Υπολλείματα Καλλιεργειών", "Σκουπιδότοποι"]].sum()
#I am printing it out to check
burned_stremmata_by_category
```

## I want to see which region had the most burned acres for each year
